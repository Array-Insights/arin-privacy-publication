```python
from arin_privacy_publication.student_t_test_statistic import StudentTTestStatistic
from arin_privacy_publication.normal_distribution_generator import NormalDistributionGenerator
data_generator = NormalDistributionGenerator()
test = StudentTTestStatistic()

import scipy
print(scipy.stats.ttest_ind([0, 1, 2], [2, 3, 4], equal_var=True))
from arin_privacy_publication.tools_privacy import compute_power
import numpy as np
sample_size = 200

list_effect_size = np.arange(-0.5, 1.0, 0.05)

list_power = []
for effect_size in list_effect_size:
    power = compute_power(test, data_generator, effect_size, sample_size, 0.05, 10)
    list_power.append(power)

import matplotlib.pyplot as plt
plt.plot(list_effect_size, list_power)
plt.xlabel("Effect size (in sdev)")
plt.ylabel("Power")
plt.show()
```
